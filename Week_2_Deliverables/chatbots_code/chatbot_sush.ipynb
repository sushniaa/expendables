```python
from datasets import load_dataset

# Load all four datasets
ds1 = load_dataset("KisanVaani/agriculture-qa-english-only")
ds2 = load_dataset("Mahesh2841/Agriculture")
ds3 = load_dataset("YuvrajSingh9886/Agriculture-Soil-QA-Pairs-Dataset")
ds4 = load_dataset("muhammad-atif-ali/agriculture-qa-english-llama-2")
ds5 = load_dataset("kshubham/agriculture_data_5k")
```

```python
print("Dataset 1:", ds1)
print("Dataset 2:", ds2)
print("Dataset 3:", ds3)
print("Dataset 4:", ds4)
print("Dataset 5:", ds5)
```

```python
def normalize_dataset(dataset, split="train"):
    texts = []
    dataset_name = dataset[split].name if hasattr(dataset[split], 'name') else "Unknown"

    # Dataset 1: KisanVaani/agriculture-qa-english-only (question, answers)
    if "question" in dataset[split].features and "answers" in dataset[split].features:
        for item in dataset[split]:
            question = item["question"]
            # Handle if answers is a list or string
            answer = item["answers"] if isinstance(item["answers"], str) else " ".join(item["answers"]) if item["answers"] else "No answer provided."
            text = f"Question: {question}\nAnswer: {answer}"
            texts.append(text)

    # Dataset 2 & 5: Mahesh2841/Agriculture, kshubham/agriculture_data_5k (instruction, input, response)
    elif "instruction" in dataset[split].features and "response" in dataset[split].features:
        for item in dataset[split]:
            instruction = item["instruction"] or ""
            input_text = item["input"] or ""
            response = item["response"] or "No response provided."
            # Combine into a single text
            text = f"Instruction: {instruction}\nInput: {input_text}\nResponse: {response}"
            texts.append(text)

    # Dataset 3: YuvrajSingh9886/Agriculture-Soil-QA-Pairs-Dataset (QUESTION.question, ANSWER, QUESTION.paragraph)
    elif "QUESTION.question" in dataset[split].features and "ANSWER" in dataset[split].features:
        for item in dataset[split]:
            question = item["QUESTION.question"] or ""
            answer = item["ANSWER"] or "No answer provided."
            paragraph = item["QUESTION.paragraph"] or ""
            # Include paragraph as context if available
            text = f"Context: {paragraph}\nQuestion: {question}\nAnswer: {answer}" if paragraph else f"Question: {question}\nAnswer: {answer}"
            texts.append(text)

    # Dataset 4: muhammad-atif-ali/agriculture-qa-english-llama-2 (formatted_text)
    elif "formatted_text" in dataset[split].features:
        for item in dataset[split]:
            formatted_text = item["formatted_text"] or ""
            # Check if it's already in Q&A format, otherwise treat as raw text
            if "Question:" in formatted_text and "Answer:" in formatted_text:
                texts.append(formatted_text)
            else:
                texts.append(f"Text: {formatted_text}")

    else:
        print(f"Unknown format for dataset {dataset_name}: {dataset[split].features}")
        return []

    print(f"Extracted {len(texts)} documents from {dataset_name}")
    return texts

# Normalize all datasets
texts1 = normalize_dataset(ds1)
texts2 = normalize_dataset(ds2)
texts3 = normalize_dataset(ds3)
texts4 = normalize_dataset(ds4)
texts5 = normalize_dataset(ds5)

# Combine all texts into a single list
all_texts = texts1 + texts2 + texts3 + texts4 + texts5
print(f"Total documents combined: {len(all_texts)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Split texts into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.create_documents(all_texts)
print(f"Total chunks after splitting: {len(docs)}")
```

```python
# Print a few sample documents
for i, doc in enumerate(docs[:3]):
    print(f"Sample Document {i+1}:")
    print(doc.page_content)
    print("-" * 50)
```

```python
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(docs, embeddings)
```

```python
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline

llm_pipeline = pipeline("text2text-generation", model="google/flan-t5-base")
llm = HuggingFacePipeline(pipeline=llm_pipeline)
```

```python
# Define the prompt
from langchain.prompts import PromptTemplate

template = """
You are an expert agriculture assistant helping Indian farmers with simple, practical, and trustworthy advice.

Your goal is to:
- Understand the farmer's question (in English or Tamil)
- Read the provided context carefully
- Answer with step-by-step solutions using simple language, no technical jargon
- Use Indian units (e.g., acres, kilograms, liters) and consider regional practices (e.g., monsoon seasons)
- Include a practical example if possible (e.g., "For example, if you're growing rice in Tamil Nadu...")
- Be encouraging and supportive
- If the question is in Tamil, reply in Tamil with clear, simple words
- If the context doesn't fully answer the question, provide a general tip if available, or politely say "I don't know based on the given information, but you can consult a local agriculture officer for more details."

📄 Context:
{context}

❓ Question from farmer:
{question}

✅ Expert Answer:

Example (for guidance):
- Question: "What crops can I grow in Kharif season?"
- Answer: During the Kharif season in India, which starts with the monsoon in June, you can grow crops like rice, maize, and cotton. Here’s how to start: First, prepare your field by plowing it well. Then, sow rice seeds at 20 kg per acre. For example, if you’re growing rice in Tamil Nadu, use the variety IR50 for better yield. Make sure to water the field regularly during the early stages.

Example (Tamil):
- Question: "கரிசல் மண்ணில் பயிரிட என்ன பயிர்கள் சிறந்தவை?"
- Answer: கரிசல் மண்ணில் பயிரிட நெல், பருத்தி, சோளம் போன்ற பயிர்கள் சிறந்தவை. முதலில், உங்கள் நிலத்தை நன்கு உழவு செய்யுங்கள். பிறகு, ஒரு ஏக்கருக்கு 20 கிலோ நெல் விதைகளை விதையுங்கள். உதாரணமாக, தமிழ்நாட்டில் IR50 வகை நெல் நல்ல பலன் தரும். பயிர் வளரும் முதல் கட்டத்தில் தண்ணீர் தொடர்ந்து பாய்ச்சுங்கள்。
"""
prompt = PromptTemplate(input_variables=["context", "question"], template=template)
```

```python
# Use ConversationalRetrievalChain instead of RetrievalQA
from langchain.chains import ConversationalRetrievalChain

chat_history = []  # Empty for now, can be used for conversation history later
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=vectorstore.as_retriever(search_kwargs={"k": 3}),
    combine_docs_chain_kwargs={"prompt": prompt},
    return_source_documents=True
)
```

```python
# Test the chain
query = "What crops can I grow in Kharif season?"
result = qa_chain({"question": query, "chat_history": chat_history})
print("Answer:", result["answer"])
print("Sources:", [doc.page_content for doc in result["source_documents"]])
```

```python
# Import required libraries
from flask import Flask, request, render_template_string

# Initialize Flask app
app = Flask(__name__)

# Initialize chat history for ConversationalRetrievalChain and Flask
chat_history = []  # For UI display
chain_history = []  # For ConversationalRetrievalChain

# HTML template for the UI
html_template = """
<!DOCTYPE html>
<html>
<head>
    <title>🌾 Farmer's Agriculture Assistant 🌾</title>
    <style>
        body {
            background-color: #f0f8e6;
            font-family: Arial, sans-serif;
            margin: 20px;
        }
        h1 {
            color: #2e7d32;
            text-align: center;
        }
        p {
            color: #4a4a4a;
            text-align: center;
        }
        .chat-container {
            border: 2px solid #2e7d32;
            border-radius: 10px;
            padding: 20px;
            margin-top: 20px;
        }
        .chat-message {
            margin-bottom: 20px;
        }
        .user-message {
            color: #1e88e5;
            font-weight: bold;
        }
        .assistant-message {
            color: #2e7d32;
        }
        form {
            text-align: center;
            margin-top: 20px;
        }
        input[type="text"] {
            width: 50%;
            padding: 10px;
            font-size: 16px;
        }
        input[type="submit"] {
            padding: 10px 20px;
            background-color: #2e7d32;
            color: white;
            border: none;
            border-radius: 5px;
            cursor: pointer;
        }
    </style>
</head>
<body>
    <h1>🌾 Farmer's Agriculture Assistant 🌾</h1>
    <p>
        Welcome! I’m here to help Indian farmers with simple, practical advice on farming.<br>
        Ask me anything about crops, fertilizers, pests, or soil in English.<br>
        <b>Examples:</b><br>
        - What crops can I grow in Kharif season?<br>
        - What is the best fertilizer for rice?<br>
        - How to control pests in wheat?
    </p>
    <form method="POST" action="/">
        <input type="text" name="query" placeholder="Ask your farming question here..." required>
        <input type="submit" value="Ask">
    </form>
    <div class="chat-container">
        {% for message in chat_history %}
            <div class="chat-message">
                <div class="{{ message.role }}-message">{{ message.role.capitalize() }}: {{ message.content }}</div>
            </div>
        {% endfor %}
    </div>
</body>
</html>
"""

# Define the route for the homepage
@app.route("/", methods=["GET", "POST"])
def home():
    global chat_history, chain_history
    
    if request.method == "POST":
        query = request.form.get("query")
        if query:
            # Add user message to chat history
            chat_history.append({"role": "user", "content": query})
            
            try:
                # Process the query using ConversationalRetrievalChain
                result = qa_chain({"question": query, "chat_history": chain_history})
                answer = result["answer"]
                sources = [doc.page_content for doc in result["source_documents"]]
                
                # Update ConversationalRetrievalChain history
                chain_history.append((query, answer))
                
                # Format the response with sources
                response = f"Answer: {answer}\n\nSources:\n" + "\n".join([f"- {source}" for source in sources])
                
                # Add assistant response to chat history
                chat_history.append({"role": "assistant", "content": response})
            
            except Exception as e:
                error_message = f"Sorry, I encountered an error: {str(e)}. Please try asking your question differently or consult a local agriculture officer."
                chat_history.append({"role": "assistant", "content": error_message})
    
    # Render the template with chat history
    return render_template_string(html_template, chat_history=chat_history)

# Run the Flask app
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)
```

```python
import gradio as gr

# Initialize chat history for ConversationalRetrievalChain
chat_history = []

# Function to handle chatbot responses (English-only)
def chatbot(query):
    global chat_history
    
    try:
        # Process the query using ConversationalRetrievalChain
        result = qa_chain({"question": query, "chat_history": chat_history})
        answer = result["answer"]
        sources = [doc.page_content for doc in result["source_documents"]]
        
        # Update chat history
        chat_history.append((query, answer))
        
        # Format the response with sources
        response = f"**Answer:** {answer}\n\n**Sources:**\n" + "\n".join([f"- {source}" for source in sources])
        return response
    
    except Exception as e:
        return f"Sorry, I encountered an error: {str(e)}. Please try asking your question differently or consult a local agriculture officer."

# Create a custom CSS for styling
css = """
    .container { 
        background-color: #f0f8e6; 
        font-family: Arial, sans-serif; 
    }
    .title { 
        color: #2e7d32; 
        font-size: 24px; 
        font-weight: bold; 
    }
    .description { 
        color: #4a4a4a; 
        font-size: 16px; 
    }
    .chatbot { 
        border: 2px solid #2e7d32; 
        border-radius: 10px; 
    }
"""

# Create the Gradio interface
with gr.Blocks(css=css) as interface:
    gr.Markdown(
        """
        <div class='title'>🌾 Farmer's Agriculture Assistant 🌾</div>
        <div class='description'>
        Welcome! I’m here to help Indian farmers with simple, practical advice on farming.  
        Ask me anything about crops, fertilizers, pests, or soil in English.  
        **Examples:**  
        - What crops can I grow in Kharif season?  
        - What is the best fertilizer for rice?  
        - How to control pests in wheat?
        </div>
        """
    )
    
    # Chat interface (no language dropdown since it's English-only)
    chat = gr.ChatInterface(
        fn=chatbot,
        title="Chat with Your Agriculture Expert",
        description="Type your question below and get step-by-step farming advice!",
        theme="soft",
        examples=[
            "What is the best fertilizer for rice?",
            "How to control pests in wheat?",
            "What crops can I grow in Kharif season?"
        ],
        textbox=gr.Textbox(placeholder="Ask your farming question here...", label="Your Question")
    )

# Launch the interface
interface.launch()
```
